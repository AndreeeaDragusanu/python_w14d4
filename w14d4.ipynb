```python
import dotenv
import os
from sqlalchemy import create_engine
import pandas as pd

dotenv.load_dotenv(override = True)

username=os.getenv("username")
password=os.getenv("password")
host=os.getenv("host")
dbname=os.getenv("dbname")

conn_string = f"mysql+pymysql://{username}:{password}@{host}/{dbname}"
db_engine = create_engine(conn_string)

query = "select * from dimproduct"

df = pd.read_sql(query, db_engine)

db_engine.dispose()

df
```

```python
"""Quanto costano in media i prodotti (colonna DealerPrice)?"""
prezzo_medio = df["DealerPrice"].mean()
round(prezzo_medio,2)
```

```python
"""Se prendiamo un quarto di tutti i prodotti, in modo che siano i più costosi, quale sarà il loro range di prezzo?"""
percentile75 = df["DealerPrice"].quantile(0.75)

maggiore_percentile75 = df[df["DealerPrice"] >= percentile75]

min = maggiore_percentile75["DealerPrice"].min()
max = maggiore_percentile75["DealerPrice"].max()
min, max
```

```python
"""• Qual è il prezzo medio per i prodotti di colore blu?"""
filtro =  df[df["Color"] == "Blue"]["DealerPrice"].mean()
round(filtro, 2)
```

```python
"""• Qual è il prezzo medio per i prodotti di colore rosso o nero?"""
filtro= df[(df["Color"] == "Red") | (df["Color"] == "Black")]["DealerPrice"].mean()
round(filtro, 2)
```

```python
"""• Qual è il prezzo massimo per i prodotti di taglia 42 e peso oltre i 10 Kg?"""

filtro = df[(df["Size"] == "42") & (df["Weight"] > 10)]["DealerPrice"].max()
round(filtro, 2)
```

```python
"""• Qual è il nome inglese e il costo di produzione (StandardCost) di tutti i prodotti di taglia 42,
peso oltre i 10 Kg e colore argento?"""
filtro = df.loc[(df["Size"] == "42") & (df["Weight"] > 10),["EnglishProductName", "StandardCost"]]
filtro
```

```python
"""Esaminiamo la differenza tra lo StandardCost e il DealerPrice degli ultimi 20 elementi del dataset: c'è un pattern? 
C'è qualche elemento che non lo segue?"""
filtro = df.tail(20)[["StandardCost", "DealerPrice"]]
filtro
```
